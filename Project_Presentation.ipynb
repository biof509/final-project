#### Functions for unpacking and processing data

```python
from IPython.display import HTML
import numpy as np
import matplotlib.pyplot as plt
import tarfile
import pickle
```

```python
# define a function to unpack the data into a dictionary
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding = 'bytes')
    return dict
```

```python
# data comes in vectors so reshape then normalize
def reshape_data(batchdata):

    # get image dimensions
    n_channels = 3
    n_imgs     = batchdata.shape[0]
    img_dim    = int(np.sqrt(batchdata.shape[1]/n_channels))
    
    # reshape vectors into images
    train_data = batchdata.reshape((n_imgs, n_channels, img_dim, img_dim)).transpose(0, 2, 3, 1)
    
    # normalize images so their values fall between 0 and 1
    train_data = (train_data / 255) - 0.5
    
    return train_data, img_dim
```

```python
# one hot encode the labels (could have just used keras.utils to_categorical)
def one_hot_encode(train_labels):
    n_imgs = len(train_labels)
    encoded = np.zeros((n_imgs, len(np.unique(train_labels))))
    
    for idx,lbl in enumerate(train_labels,0):
        encoded[idx][lbl] = 1
    
    return encoded
```

#### Unpack train and test data and combine train batches

```python
# training
batchpath  = os.path.join(projectpath, 'cifar-10-batches-py')

for set in range(1,6):
    if set == 1:
        dirname             = os.path.join(batchpath, 'data_batch_' + str(set))
        batch               = unpickle(dirname)
        train_data, img_dim = reshape_data(batch[b'data'])        
        train_labels        = one_hot_encode(batch[b'labels'])
    else:
        dirname               = os.path.join(batchpath, 'data_batch_' + str(set))
        batch                 = unpickle(dirname)
        train_data2, img_dim2 = reshape_data(batch[b'data'])
        train_labels2         = one_hot_encode(batch[b'labels'])
        train_data            = np.concatenate((train_data, train_data2))
        train_labels          = np.concatenate((train_labels, train_labels2))
```

```python
# testing
batchpath          = os.path.join(projectpath, 'cifar-10-batches-py')
dirname            = os.path.join(batchpath, 'test_batch')
testbatch          = unpickle(dirname)
test_data, img_dim = reshape_data(testbatch[b'data'])        
test_labels        = one_hot_encode(testbatch[b'labels'])
```

## 7) The Model

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.models    import Sequential
from keras.layers    import Dense, Conv2D, Flatten, Dropout, MaxPool2D, BatchNormalization, Activation
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.models    import load_model
```

<font size="3.7">Convolution</font>

<video width="320" height="240" controls src="convolution_video.mp4" />

<font size="3.7">MaxPooling</font>
- Downsamples input by taking the maximum value in the window

<font size="3.7">Dropout</font>
- Ignoring a random subset of units during the training phase for a particular forward/backward pass

<font size="3.7">Batch Normalization</font>
- Normalize input values of the same unit from different images in one mini batch

```python
# to see output of running the model, see Data_Processing notebook
model = Sequential()

# =========== Convolution Layer 1 ===========
model.add(Conv2D(32, 
                 kernel_size = 3, 
                 activation  = 'relu', 
                 padding     = 'same', 
                 input_shape = (img_dim,
                                img_dim,
                                3)))
model.add(MaxPool2D(2))
model.add(Dropout(0.25))

# =========== Convolution Layer 2 ===========
model.add(Conv2D(32,
                 kernel_size = 5, 
                 activation  = 'relu',
                 padding     = 'same'))
model.add(MaxPool2D(2))
model.add(Dropout(0.25))

# =========== Convolution Layer 3 ===========
model.add(Conv2D(64,
                 kernel_size = 3))
model.add(BatchNormalization())
model.add(Activation('relu'))

# =========== Convolution Layer 4 ===========
model.add(Conv2D(64, 
                 activation  = 'relu',
                 kernel_size = 3))
model.add(MaxPool2D(2))
model.add(Flatten())

# =========== Fully Connected Layers ===========
model.add(Dense  (64,  activation = 'relu'))
model.add(Dense  (128))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense  (64,  activation = 'relu'))
model.add(Dense  (10,  activation = 'softmax'))

# =========== Training Parameters ===========
early_stopping_monitor = EarlyStopping(patience = 15,
                                       restore_best_weights = True)

model.compile(optimizer = 'adam', 
              loss      = 'categorical_crossentropy', 
              metrics   = ['accuracy'])

# =========== Training Model ===========
training = model.fit(train_data, 
                     train_labels, 
                     validation_split = 0.2, 
                     epochs           = 100, 
                     batch_size       = 125,
                     callbacks        = [early_stopping_monitor])

#model.save('model_saved9.h5')

plt.plot(training.history['loss'])
plt.plot(training.history['val_loss'])
plt.show()
```

```python
def convolution(kernel, image):
    img_dim     = image.shape[0]
    channels    = image.shape[2]
    kernel_size = kernel.shape[0]
    output_size = img_dim-(kernel_size-1)
    conv        = np.zeros((output_size,output_size))
    
    for ii in range(output_size):
        for jj in range(output_size):
            
            window = image[ii:ii + kernel_size, 
                           jj:jj + kernel_size]
            
            conv[ii,jj] = np.sum(window * kernel)
            
    return conv

# code to generate image below is in Data_Processing
```

### Model Performance per Category

```python
pred_cat      = np.sum(predictions, 
                       axis=0)
corr_pred_cat = np.sum(predictions * test_labels,
                       axis=0)
true_cat      = np.sum(test_labels, 
                       axis=0)
prob_cat      = corr_pred_cat/true_cat

x = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
     'dog', 'frog', 'horse', 'ship', 'truck']

fig = plt.figure()
ax  = fig.add_subplot(111)
ax.bar(x, prob_cat); 
ax.tick_params(axis='both', 
               colors='black') 
ax.tick_params(axis='x',
               labelrotation=45)
```
